```python
# CELL BARU (Copy ini)

from solders.keypair import Keypair
from solana.rpc.api import Client

# 1. Setup Koneksi
client = Client("https://api.devnet.solana.com")

# 2. Buat Dompet Baru (Wallet)
wallet_baru = Keypair()

print("📦 DOMPET BARU BERHASIL DIBUAT!")
print(f"📍 Alamat Dompet (Public Key): {wallet_baru.pubkey()}")
print("⚠️ INGAT: Jangan share Secret Key di tempat umum, tapi untuk hackathon ini tidak masalah.")

# 3. Minta Uang Gratis (Airdrop 2 SOL)
print(f"\n💰 Meminta 2 SOL gratis untuk dompet ini... (Tunggu sebentar)")
try:
    # Meminta 2 SOL (1 SOL = 1,000,000,000 Lamports)
    airdrop = client.request_airdrop(wallet_baru.pubkey(), 2 * 10**9)
    print(f"✅ Berhasil! Transaksi sedang diproses. Signature: {airdrop.value}")
except Exception as e:
    print(f"❌ Error: {e}")
```

```python
# CELL BERIKUTNYA (Copy ini setelah上面的 cell selesai jalan)

from solana.rpc.commitment import Confirmed
import time

time.sleep(5) # Tunggu 5 detik agar transaksi masuk

# Cek saldo
saldo = client.get_balance(wallet_baru.pubkey(), commitment=Confirmed)

# Konversi dari Lamports ke SOL (bagi 1 miliar)
saldo_solar = saldo.value / 1_000_000_000

print(f"💎 Saldo Dompet Kamu: {saldo_solar} SOL")
```
